## **Atividade 1**

*Utilizando os algorítmos de conversão vistos nas aulas de Cálculo Numérico, crie uma função chamada* **convert10to2** *em Python, que solicita do usuário um número real positivo em base decimal e o converte para base binária com, no máximo, 10 casas decimais. Sua função deve aceitar somente numeros da forma* **n . d** *ou* **n , d**, *onde* **n** *é a parte inteira do número e* **d** *é a parte decimal. Por exemplo, os números 12.45 e 15,10 são aceitos, mas 12/7 ou -123,23 ou -329.13, etc. devem ser recusados com a mensagem "Número Inválido".*




```python
def convert10to2():
    import re

    # Regex para validar o número inserido pelo usuário
    regex = r'^\d+([.,]\d+)?$'

    # Solicita o número do usuário
    num_str = input("Digite um número real positivo em base decimal (ex: 12.45 ou 15,10): ")

    # Verifica se o número é válido
    if not re.match(regex, num_str):
        print("Número Inválido")
        return

    # Substitui a vírgula por ponto, se necessário
    num_str = num_str.replace(',', '.')

    # Separa a parte inteira e a parte fracionária
    if '.' in num_str:
        n_str, d_str = num_str.split('.')
        inteiro = int(n_str)
        decimal = float('0.' + d_str)
    else:
        inteiro = int(num_str)
        decimal = 0.0

    # Função para converter a parte inteira para binário
    def inteiro_para_binario(num):
        if num == 0:
            return '0'
        binario = ''
        while num > 0:
            binario = str(num % 2) + binario
            num = num // 2
        return binario

    # Função para converter a parte fracionária para binário
    def fracao_para_binario(fracao):
        binario = ''
        contador = 0
        while fracao > 0 and contador < 10:
            fracao = fracao * 2
            bit = int(fracao)
            binario += str(bit)
            fracao -= bit
            contador += 1
        return binario

    # Converte a parte inteira
    parte_inteira_bin = inteiro_para_binario(inteiro)

    # Converte a parte fracionária
    parte_fracionaria_bin = fracao_para_binario(decimal)

    # Combina as partes
    if decimal == 0:
        resultado = parte_inteira_bin
    else:
        resultado = parte_inteira_bin + '.' + parte_fracionaria_bin

    print(f'O número {num_str} em binário é: {resultado}')


convert10to2()
```

**Dicas**
1.   Considere separadamente os casos em que o número a ser convertivo é inteiro (isto é, a parte fracionária é zero) e quando não é inteiro (parte fracionária diferente de zero);

2.   Separe a parte inteira da parte fracionária do número para aplicar os algorítmos em cada caso e depois combine os resultados;

3.   Observe a ordem em que as operações são feitas em cada algorítmo visto nas aulas.

```python
# @title **Código de teste para Atividade 1**
try:
  assert convert10to2("95") == 1011111
  assert convert10to2("123.234") == 1111011.0011101111
  assert convert10to2("37.52") == 100101.10000101
  assert convert10to2("0.625") == 0.101
  assert convert10to2("-21,12") == "Número Inválido"
  assert convert10to2("12/7") == "Número Inválido"
  assert convert10to2("23.43,2") == "Número Inválido"
  assert convert10to2("20;30") == "Número Inválido"

  print("Passou em todos os testes! :D")

except AssertionError:
  print( "Falhou em um ou mais testes... :(")
  raise
```

# **Atividade 2**

*Utilizando os algorítmos de conversão vistos nas aulas de Cálculo Numérico, crie uma função chamada* **convert2to10** *em Python, que solicita do usuário um número positivo em base binária e o converte para base decimal. Sua função deve aceitar somente numeros da forma* **n . d**, *onde* **n** *é a parte inteira do número e* **d** *é a parte fracionária. Se os números não estiverem nesse formato ou seus dígitos não forem binários a função deve recusá-los, acusando um erro do tipo "Número inserido não é binário" ou "Formatação numérica inválida", conforme o caso.*


```python
def convert2to10():
    import re

    # Regex para validar se o número está no formato correto
    regex = r'^[01]+(\.[01]+)?$'

    # Solicita o número binário do usuário
    bin_str = input("Digite um número binário (ex: 1101.101): ")

    # Verifica se o número está no formato correto
    if not re.match(regex, bin_str):
        if not all(c in '01.' for c in bin_str):
            print("Número inserido não é binário")
        else:
            print("Formatação numérica inválida")
        return

    # Separa a parte inteira e a parte fracionária
    if '.' in bin_str:
        n_str, d_str = bin_str.split('.')
    else:
        n_str = bin_str
        d_str = ''

    # Função para converter a parte inteira para decimal
    def inteiro_para_decimal(num_str):
        decimal = 0
        for i, digit in enumerate(reversed(num_str)):
            decimal += int(digit) * (2 ** i)
        return decimal

    # Função para converter a parte fracionária para decimal
    def fracao_para_decimal(frac_str):
        decimal = 0
        for i, digit in enumerate(frac_str, start=1):
            decimal += int(digit) * (2 ** -i)
        return decimal

    # Converte a parte inteira para decimal
    parte_inteira_dec = inteiro_para_decimal(n_str)

    # Converte a parte fracionária
    if d_str:
        parte_fracionaria_dec = fracao_para_decimal(d_str)
    else:
        parte_fracionaria_dec = 0.0

    # Combina as partes
    resultado = parte_inteira_dec + parte_fracionaria_dec

    print(f'O número binário {bin_str} em decimal é: {resultado}')


convert2to10()
```

**Dicas**
1.   Primeiro verique se o número está no formato correto usando, por exemplo, a função isnumeric() em combinação com outros condicionais.

2.   Considere separadamente os casos em que a parte fracionária do número é zero e diferente de zero;

3.   Separe a parte inteira da parte fracionária do número para aplicar os algorítmos em cada caso e depois combine os resultados;

4.   Observe a ordem em que as operações são feitas em cada algorítmo.

```python
# @title **Código de teste para Atividade 2**
try:
  assert convert2to10("100101.1000010100") == 37.51953125
  assert convert2to10("0.10011") == 0.59375
  assert convert2to10("10100011") == 163
  assert convert2to10("11011.111") == 27.875
  assert convert2to10("-101.111") == "Formatação numérica inválida."
  assert convert2to10("120.111") == "Número inserido não é binário."
  assert convert2to10("23.43,2") == "Número inserido não é binário."
  assert convert2to10("-101.1101,101") == "Formatação numérica inválida."

  print("Passou em todos os testes! :D")

except AssertionError:
  print( "Falhou em um ou mais testes... :(")
  raise
```
